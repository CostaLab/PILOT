```r
library(PMCMRplus)
library(tsutils)
```

# ARI

```r
# Read the CSV file
data <- read.csv('/data/mu0611151/data/mask/PILOT_revise/Friedman-Nemenyi/Data/ARI.csv')
# Extract the matrix from the data frame
mat <- as.matrix(data)
# Set row names
rownames(mat) <- 1:nrow(mat)
# Set column names
colnames(mat) <- c("PhEMD", "PseudoBulk", "PILOT", "Proportions")
```

### Friedman-Nemenyi

```r
results_ari<-nemenyi(-mat,plottype="vmcb")
```

```r
results_ari
```

#### Nemenyi_intervals

```r
results_ari$intervals
```
