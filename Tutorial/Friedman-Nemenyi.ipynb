# ARI

```python
# Read the CSV file
data <- read.csv('Datasets/ARI.csv')
# Extract the matrix from the data frame
mat <- as.matrix(data)
# Set row names
rownames(mat) <- 1:nrow(mat)
# Set column names
colnames(mat) <- c("PhEMD", "PseudoBulk", "PILOT", "Proportions")
```

### Friedman-Nemenyi

```python
results_ari<-nemenyi(-mat,plottype="vmcb")
```

```python
results_ari
```

#### Nemenyi_intervals

```python
results_ari$intervals
```
